```python
# Ignore the other values in churn status like Joined as it is very small dataset. Leave it to Null! 

query = """
SELECT `customer`.`customer_id`,
    `customer`.`gender`,
    `customer`.`age`,
    case when `customer`.`senior_citizen` = 'Yes' then 1 else 0 end as `senior_citizen`,
    case when `customer`.`married` = 'Yes' then 1 else 0 end as `married`,
    `customer`.`num_dependents`,
    `customer`.`zip_code`,
    `city`.`city`,
    `city`.`latitutde`,
    `city`.`longitude`,
    `city`.`population`,
    case when `churn_status`.`status` = 'Stayed' then 0  when `churn_status`.`status` = 'Churned' then 1 end as `churn_status`,
    case when `churn_status`.`churn_label` = 'Yes' then 1 when `churn_status`.`churn_label` = 'No' then 0 end as `churn_label` ,
    `churn_status`.`churn_category`,
    `churn_status`.`churn_reason`,
	`account`.`account_id`,
    `account`.`tenure_months`,
    `account`.`num_referrals`,
    case when `account`.`has_internet_service` = 'Yes' then 1 when `account`.`has_internet_service` = 'No' then 0 end as `has_internet_service`,
    `account`.`internet_type`,
    case when `account`.`has_unlimited_data` = 'Yes' then 1 else 0 end as `has_unlimited_data`,
    case when `account`.`has_phone_service` = 'Yes' then 1 else 0 end as `has_phone_service`,
    case when `account`.`has_multiple_lines` = 'Yes' then 1 else 0 end as `has_multiple_lines`,
    case when `account`.`has_premium_tech_support` = 'Yes' then 1 else 0 end as `has_premium_tech_support`,
    case when `account`.`has_online_security` = 'Yes' then 1 else 0 end as `has_online_security`,
    case when `account`.`has_online_backup` = 'Yes' then 1 else 0 end as `has_online_backup`,
    case when `account`.`has_device_protection` = 'Yes' then 1 else 0 end as `has_device_protection`,
    `account`.`contract_type`,
    case when `account`.`paperless_billing` = 'Yes' then 1 else 0 end as `paperless_billing`,
    `account`.`payment_method`,
    `account_usage`.`avg_long_distance_fee_monthly`,
    `account_usage`.`total_long_distance_fee`,
    `account_usage`.`avg_gb_download_monthly`,
    case when `account_usage`.`stream_tv` = 'Yes' then 1 else 0 end as `stream_tv`,
    case when `account_usage`.`stream_movie` = 'Yes' then 1 else 0 end as `stream_movie`,
    case when `account_usage`.`stream_music` = 'Yes' then 1 else 0 end as `stream_music`,
    `account_usage`.`total_monthly_fee`,
    `account_usage`.`total_charges_quarter`,
    `account_usage`.`total_refunds`
FROM `ai300_capstone`.`customer`
LEFT JOIN `ai300_capstone`.`city` ON `customer`.`zip_code` = `city`.`zip_code`
LEFT JOIN `ai300_capstone`.`churn_status` ON `churn_status`.`customer_id` = `customer`.`customer_id`
LEFT JOIN `ai300_capstone`.`account` ON `account`.`customer_id` = `customer`.`customer_id`
LEFT JOIN `ai300_capstone`.`account_usage` on `account_usage`.`account_id` = `account`.`account_id`
;
"""
combined_df = pd.DataFrame(database_conn.connect().execute(text(query)))
combined_df.head()
```

```python
# replace all the blank values with null
combined_df = combined_df.replace(r'^\s*$', np.nan, regex=True)
combined_df.info()
```

```python
# We will need only rows with churned label yes/no
combined_df = combined_df.dropna(axis=0,subset=['churn_label'])
combined_df.info()
```

```python
# rest of the churn columns that are blank are not going to be features for model anyway. ignore those nulls!  
# just for reporting needs - we label them as 'not churned'
fill_values = {"churn_category": "not_churned", "churn_reason": "not_churned"}
combined_df = combined_df.fillna(value=fill_values)
combined_df.info()
```

```python
combined_df['churn_label'] = combined_df['churn_label'].astype(int)
combined_df.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
import statistics
import xgboost as xgb

# Split data into predictors X and output Y
features = ['gender','age','senior_citizen','married','num_dependents',
'zip_code','tenure_months','num_referrals',
'contract_type_Month-to-Month','contract_type_One Year','contract_type_Two Year',
#'has_internet_service','has_unlimited_data','has_phone_service','has_multiple_lines','has_premium_tech_support',
#'has_online_security','has_online_backup','has_device_protection',
#'paperless_billing',
#'avg_long_distance_fee_monthly','total_long_distance_fee','avg_gb_download_monthly',
#'stream_tv','stream_movie','stream_music',
'total_monthly_fee','total_charges_quarter']
#'latitutde','longitude','population','total_refunds',
## below are categorical values removed for now
#'internet_type','contract_type','payment_method']
X = combined_df[features] 
#print(X)
y = combined_df['churn_label']

# Split the data into ten set
kf = KFold(n_splits=9)

# List of accuracy for each fold
k_fold_accuracy = []

# Iterate through each fold and calculate the RMSE for each fold
for train_index, test_index in kf.split(X):
    
    # Extract the training and test data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     ############
   
    # Instantiate the model object
    xbg_model = xgb.XGBClassifier(random_state=0,           
                                    learning_rate =0.1,
                                    n_estimators=88,
                                    max_depth=18,
                                    min_child_weight=6,
                                    gamma=0,
                                    subsample=0.8,
                                    colsample_bytree=0.8,
                                    objective= 'binary:logistic',
                                    nthread=4,
                                    scale_pos_weight=1,
                                    seed=27,
                                    enable_categorical=True)

    #le = LabelEncoder()
    #y_train = le.fit_transform(y_train)

    # Fit the model with the training data
    xbg_model.fit(X_train, y_train)

    # predict the target on the test dataset
    y_predict = xbg_model.predict(X_test)

    #predict probabilities and 
    y_pred_proba = xbg_model.predict_proba(X_test)[:,1]
    ## print('\nPrediction on test data', y_predict) 

    auc_score = metrics.roc_auc_score(y_test, y_pred_proba)
    print('AUC:', auc_score)

    ## best would be to combine all AUC charts in one and show mean value. try next time when you have time! 
    ## https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html
    #############################
    ## AUC curve
    #############################
    # Calculate the (1) false positive rate, (2) true positive rate, and (3) thresholds
    # passing pos_label as the eror stated without having this parameter: 
    # y_true takes value in {'0', '1'} and pos_label is not specified: either make y_true take value in {0, 1} or {-1, 1} or pass pos_label explicitly.
    y_pred_probabilities_formatted = list(y_pred_proba) # converting to list
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label=1)

    # Plotting the chart
    fig = px.area(
        x=fpr, y=tpr,
        title=f'ROC Curve (AUC={auc(fpr, tpr):.4f})',
        labels=dict(x='False Positive Rate', y='True Positive Rate'),
        width=700, height=500
    )

    # This part is just for formatting & adding the dash-line 
    fig.add_shape(
        type='line', line=dict(dash='dash'),
        x0=0, x1=1, y0=0, y1=1
    )
    fig.update_yaxes(scaleanchor="x", scaleratio=1)
    fig.update_xaxes(constrain='domain')


    fig.show()

    # Accuracy Score on test dataset
    #Calculate Accuracy for the fold and append it
    accuracy = metrics.accuracy_score(y_test, y_predict)
    k_fold_accuracy.append(accuracy)
    ############
    
print('The accuracy for each fold is:', k_fold_accuracy)
print('The average accuracy for XGB is:', statistics.mean(k_fold_accuracy))
```
